# ***1.Importing Libraries***

```python
from selenium import webdriver
import time
import pandas as pd
from selenium.webdriver.chrome.options import Options
```

## 2.Taking Input from User

```python
month= input("Enter the month: ")
year= input("Enter the year: ")
```

### 3.Data Scrapping Using Geckodriver and Selenium

```python
options= Options()
path='C:\\Users\\ninja\\geckodriver.exe'
#options.binary_location = "C:/Program Files/Google/Chrome/Application/chrome.exe"
driver= webdriver.Firefox(executable_path=path)
```

```python
driver.get("https://www.accuweather.com/en/in/kolkata/206690/"+month+"-weather/206690?year="+year+"&view=list")
time.sleep(10)
```

```python
data= driver.find_elements_by_class_name("high") #temp high
temp_high=[]
for i in data:
    t=i.get_attribute('textContent')
    
    t=t.replace("\t","")
    t=t.replace("\n","")
    
    temp_high.append(t[:-1])
    print(temp_high)
```

```python
  print(temp_high)
```

```python
data= driver.find_elements_by_class_name("low") #temp low
temp_low=[]
for i in data:
    t=i.get_attribute('textContent')
    
    t=t.replace("\t","")
    t=t.replace("\n","")
    
    temp_low.append(t[:-1])
    print(temp_low)
```

```python
  print(temp_low)
```

```python
data= driver.find_elements_by_class_name("date") #date
date=[]
for i in data:
    t=i.get_attribute('textContent')
    
    t=t.replace("\t","")
    t=t.replace("\n","")
    
    date.append(t)
```

```python
  print(date)
```

```python
k=0

idx=date.index('1')
date=date[idx:]     #slicing is done
temp_high=temp_high[idx:]
temp_low=temp_low[idx:]

for i in range(1,len(date)):
    if(int(date[i-1])>int(date[i])):
       idx2=i
       k=1
       break

if(k==1):
    date=date[:idx2]  #Reverse slicing
    temp_high=temp_high[:idx2] 
    temp_low=temp_low[:idx2]
       
print(date)
print(temp_high)
print(temp_low)
```

```python
d={'Date':date,'Temparature(High[in degrees])':temp_high,'Temparature(Low[in degrees])':temp_low}
df=pd.DataFrame(d)
print(df)
```
